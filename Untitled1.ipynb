**Ödev 4**

*Fibonacci*


```
def fib(n):
  a=0
  b=1

  if n==1:
    print(a)
  else:
    print(a)
    print(b)
    
    for i in range(2,n):
      c=a+b
      a=b
      b=c
      if c<=4000000:
        if c %2==0:
          print(c)

fib(4000000)
```
